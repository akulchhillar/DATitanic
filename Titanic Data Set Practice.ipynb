```python
def male_female_child(passenger):
    # Take the Age and Sex
    age,sex = passenger
    # Compare the age, otherwise leave the sex
    if age < 16:
        return 'child'
    else:
        return sex
    

# We'll define a new column called 'person', remember to specify axis=1 for columns and not index
titanic_df['Person'] = titanic_df[['Age','Sex']].apply(male_female_child,axis=1)
```

```python
titanic_df
```

```python
sns.countplot(data=titanic_df,x="Pclass",hue="Person") 
```

In all classes Males were highest and children were least

```python
titanic_df["Status"] = titanic_df["Survived"]
titanic_df["Status"].loc[titanic_df["Status"]==1] = "Yes"
titanic_df["Status"].loc[titanic_df["Status"]==0] = "No"
```

```python
sns.countplot(data=titanic_df,x="Status",hue="Sex")
plt.xlabel("Whether Survived")
```

```python
titanic_df["Alone"] = titanic_df["SibSp"] + titanic_df["Parch"]
titanic_df["Alone"].loc[titanic_df["Alone"]>0] = "With Family"
titanic_df["Alone"].loc[titanic_df["Alone"]==0] = "Alone"
```

```python
sns.factorplot(x="Alone",data=titanic_df,y="Survived")
```

```python
# First we'll drop the NaN values and create a new object, deck
deck = titanic_df['Cabin'].dropna()

levels = []

# Loop to grab first letter
for level in deck:
    levels.append(level[0])    
```

```python
cabin_df = pd.DataFrame(levels)
cabin_df.columns = ['Cabin']
sns.countplot(x='Cabin',data=cabin_df,palette='winter_d')
    
```

People in C deck were maximum

```python
titanic_df["Cabin"] = cabin_df["Cabin"]
```

```python
sns.factorplot(data=titanic_df,y="Survived",x="Cabin")
```

Most of T deck people survived the event.
